```python
from models.utils.check_corrupted import is_corrupted, create_coordinates
import os
import os.path as osp
import numpy as np
# from analysis.exploratory_data_analysis.utils.visual_utils import plot_patch_data
from matplotlib import pyplot as plt
```

```python
data_dir_path = r'C:\Users\kuipe\OneDrive\Bureaublad\Epoch\forestbiomass\data\test'
patch_names_path = [osp.join(data_dir_path, file) for file in os.listdir(data_dir_path)]
```

```python
def plot_patch_data(patch_data):
    plt.imshow(patch_data)
    plt.colorbar()
    plt.show()
```

```python
for patch in patch_names_path:
    for month in range(12):
        bands_s1 = [osp.join(patch + f'/{month}/S1/', file) for file in os.listdir(patch+ f'/{month}/S1')]
        for band_s1 in bands_s1:
            loaded_s1 = np.load(band_s1)
            if is_corrupted(loaded_s1):
                print(band_s1)
                plot_patch_data(loaded_s1)
        try:
            bands_s2 = [osp.join(patch + f'/{month}/S2/', file) for file in os.listdir(patch+ f'/{month}/S2')]
            for band_s2 in bands_s2:
                loaded_s2 = np.load(band_s2)
                if is_corrupted(loaded_s2):
                    print(band_s2)
                    plot_patch_data(loaded_s2)
        except:
            continue
```
